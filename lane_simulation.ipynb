```python
# importing things for setting up
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp

# Vehicle parameters
L = 2.5       # wheelbase [m]
dt = 0.1      # timestep [s]
N_steps = 80  # simulation steps
v = 5.0       # velocity [m/s]

# Horizon for MPC
N_horizon = 10  

# Reference paths
straight_path = [(k*dt*v, 0) for k in range(N_steps)]
curved_path   = [(k*dt*v, 2*np.sin(0.05*k*dt*v)) for k in range(N_steps)]

# Conditions to test
conditions = {
    "Initial offset": (0, 2.0, 0),   # start 2 m off center
    "Initial heading": (0, 0.0, 0.3),# start pointing at angle
    "Curved road": (0, 0.0, 0)       # road has curve
}
```

```python
# pid controller simulation
def simulate_pid(x0, y0, theta0, ref_path, Kp=0.2):
    x, y, theta = x0, y0, theta0
    states, deltas = [], []

    for k in range(N_steps):
        ref_x, ref_y = ref_path[k]
        error = ref_y - y
        delta = Kp * error
        delta = np.clip(delta, -0.5, 0.5)

        # Update vehicle
        x = x + v*np.cos(theta)*dt
        y = y + v*np.sin(theta)*dt
        theta = theta + (v/L)*np.tan(delta)*dt

        states.append([x, y, theta])
        deltas.append(delta)

    return np.array(states), np.array(deltas)
```

```python
# mpc controller simulatioin
def simulate_mpc(x0, y0, theta0, ref_path):
    x, y, theta = x0, y0, theta0
    states, deltas = [], []

    for k in range(N_steps):
        ref_x, ref_y = ref_path[k]

        # Optimization variables
        x_var = cp.Variable(N_horizon+1)
        y_var = cp.Variable(N_horizon+1)
        theta_var = cp.Variable(N_horizon+1)
        delta_var = cp.Variable(N_horizon)

        constraints = [x_var[0]==x, y_var[0]==y, theta_var[0]==theta]
        cost = 0

        for t in range(N_horizon):
            cost += cp.square(y_var[t] - ref_y) + 0.1*cp.square(delta_var[t])
            constraints += [x_var[t+1] == x_var[t] + v*dt]
            constraints += [y_var[t+1] == y_var[t] + v*theta_var[t]*dt]
            constraints += [theta_var[t+1] == theta_var[t] + (v/L)*delta_var[t]*dt]

            constraints += [cp.abs(delta_var[t]) <= 0.5]

        prob = cp.Problem(cp.Minimize(cost), constraints)
        prob.solve(solver=cp.OSQP, warm_start=True)

        delta = delta_var.value[0] if delta_var.value is not None else 0.0

        # Update vehicle
        x = x + v*np.cos(theta)*dt
        y = y + v*np.sin(theta)*dt
        theta = theta + (v/L)*np.tan(delta)*dt

        states.append([x, y, theta])
        deltas.append(delta)

    return np.array(states), np.array(deltas)
```

```python
# initial offset(cars strats 2m away from the lane)
condition = "Initial offset"
x0, y0, theta0 = conditions[condition]
path = straight_path

states_pid, deltas_pid = simulate_pid(x0, y0, theta0, path)
states_mpc, deltas_mpc = simulate_mpc(x0, y0, theta0, path)

# Trajectory
plt.figure(figsize=(6,5))
plt.plot(np.array(path)[:,0], np.array(path)[:,1], 'k--', label="Reference")
plt.plot(states_pid[:,0], states_pid[:,1], 'r-', label="PID")
plt.plot(states_mpc[:,0], states_mpc[:,1], 'b-', label="MPC")
plt.title(f"Trajectory - {condition}")
plt.xlabel("x [m]"); plt.ylabel("y [m]")
plt.legend(); plt.grid()
plt.show()

# Steering
t = np.arange(N_steps)*dt
plt.figure(figsize=(6,5))
plt.plot(t, np.degrees(deltas_pid), 'r-', label="PID")
plt.plot(t, np.degrees(deltas_mpc), 'b-', label="MPC")
plt.title(f"Steering Input - {condition}")
plt.xlabel("Time [s]"); plt.ylabel("Steering [deg]")
plt.legend(); plt.grid()
plt.show()
```

```python
# In case of this initial offset
# PID:- Overreacts causes zigzag corrections, steering also zigzag
# MPC:- Smoothly brings the car to centre line, steering is still somewhat smoother
```

```python
# initial heading(car starts angled to the road)
condition = "Initial heading"
x0, y0, theta0 = conditions[condition]
path = straight_path

states_pid, deltas_pid = simulate_pid(x0, y0, theta0, path)
states_mpc, deltas_mpc = simulate_mpc(x0, y0, theta0, path)

# Trajectory
plt.figure(figsize=(6,5))
plt.plot(np.array(path)[:,0], np.array(path)[:,1], 'k--', label="Reference")
plt.plot(states_pid[:,0], states_pid[:,1], 'r-', label="PID")
plt.plot(states_mpc[:,0], states_mpc[:,1], 'b-', label="MPC")
plt.title(f"Trajectory - {condition}")
plt.xlabel("x [m]"); plt.ylabel("y [m]")
plt.legend(); plt.grid()
plt.show()

# Steering
t = np.arange(N_steps)*dt
plt.figure(figsize=(6,5))
plt.plot(t, np.degrees(deltas_pid), 'r-', label="PID")
plt.plot(t, np.degrees(deltas_mpc), 'b-', label="MPC")
plt.title(f"Steering Input - {condition}")
plt.xlabel("Time [s]"); plt.ylabel("Steering [deg]")
plt.legend(); plt.grid()
plt.show()
```

```python
# in above angles road case
# PID:- corrects slowly often with overshoot, starts with big steering angle to correct wrong heading and then overshoots
# MPC:- stabilizes faster, starts with a great angle but then balanced steering correction.
```

```python
# curved road(reference path is a sine wave)
condition = "Curved road"
x0, y0, theta0 = conditions[condition]
path = curved_path

states_pid, deltas_pid = simulate_pid(x0, y0, theta0, path)
states_mpc, deltas_mpc = simulate_mpc(x0, y0, theta0, path)

# Trajectory
plt.figure(figsize=(6,5))
plt.plot(np.array(path)[:,0], np.array(path)[:,1], 'k--', label="Reference")
plt.plot(states_pid[:,0], states_pid[:,1], 'r-', label="PID")
plt.plot(states_mpc[:,0], states_mpc[:,1], 'b-', label="MPC")
plt.title(f"Trajectory - {condition}")
plt.xlabel("x [m]"); plt.ylabel("y [m]")
plt.legend(); plt.grid()
plt.show()

# Steering
t = np.arange(N_steps)*dt
plt.figure(figsize=(6,5))
plt.plot(t, np.degrees(deltas_pid), 'r-', label="PID")
plt.plot(t, np.degrees(deltas_mpc), 'b-', label="MPC")
plt.title(f"Steering Input - {condition}")
plt.xlabel("Time [s]"); plt.ylabel("Steering [deg]")
plt.legend(); plt.grid()
plt.show()
```

```python
# in curved road
# PID:- do not follow curve as such, more oscillations, constantly moving left-right as it reacts late to the curve
# MPC:- predicts turns and keeps the car aligned, follows the curve smoothly
```
